Q1: What is hypothesis testing in statistics?
- A: Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves formulating a null hypothesis (H‚ÇÄ) and an alternative hypothesis (H‚ÇÅ) and using statistical tests to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative.

Q2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
- A: The null hypothesis (H‚ÇÄ) is a statement that assumes no effect or no difference in a population. It serves as the default assumption. The alternative hypothesis (H‚ÇÅ) contradicts the null and represents the effect or difference that the researcher wants to test. The goal of hypothesis testing is to determine whether there is enough statistical evidence to reject H‚ÇÄ in favor of H‚ÇÅ.

Q3: What is the significance level in hypothesis testing, and why is it important?
- A: The significance level (denoted as Œ±) is the probability of rejecting the null hypothesis when it is actually true. It is usually set at 0.05 (5%), meaning there is a 5% risk of making a Type 1 error. A lower Œ± reduces the risk of incorrect rejections but may increase the chance of failing to detect a true effect.

Q4: What does a P-value represent in hypothesis testing?
- A: A P-value is the probability of obtaining test results at least as extreme as the observed data, assuming that the null hypothesis is true. It quantifies the strength of the evidence against H‚ÇÄ.

Q5: How do you interpret the P-value in hypothesis testing?
- A: - If the P-value is less than or equal to the significance level (Œ±), reject the null hypothesis (H‚ÇÄ), suggesting strong evidence against it.

- If the P-value is greater than Œ±, fail to reject H‚ÇÄ, meaning there is not enough evidence to support the alternative hypothesis.

Q6: What are Type 1 and Type 2 errors in hypothesis testing?
- A: - Type 1 Error (False Positive): Rejecting a true null hypothesis. The probability of this occurring is the significance level (Œ±).

- Type 2 Error (False Negative): Failing to reject a false null hypothesis. The probability of this occurring is Œ≤ (related to statistical power).

Q7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- A: - One-tailed test: Tests for an effect in only one direction (e.g., greater than or less than).

- Two-tailed test: Tests for an effect in both directions (e.g., not equal to). A two-tailed test is more conservative since it considers deviations in both directions.

Q8: What is the Z-test, and when is it used in hypothesis testing?
- A: A Z-test is used when the population variance is known and the sample size is large (n ‚â• 30). It tests hypotheses about population means or proportions by comparing sample data to the population using the Z-score.

Q10: What is the T-distribution, and when should it be used instead of the normal distribution?
- A: The T-distribution is a probability distribution used when the sample size is small (n < 30) and the population standard deviation is unknown. It has heavier tails than the normal distribution, making it more appropriate for estimating population parameters with small samples.

Q11: What is the difference between a Z-test and a T-test?
- A: Z-test is used when the population variance is known and the sample size is large.
- A T-test is used when the population variance is unknown and the sample size is small. The T-test accounts for additional variability in small samples.

Q12: What is the T-test, and how is it used in hypothesis testing?
- A: The T-test is a statistical test used to compare the means of two groups and determine if they are significantly different. It is commonly used in small sample sizes where the population standard deviation is unknown.

Q13: What is the relationship between Z-test and T-test in hypothesis testing?
- A: Both tests assess the difference between sample and population means. The
- Z-test is used for larger samples with known variance, while the T-test is used for smaller samples with unknown variance.

Q14: What is a confidence interval, and how is it used to interpret statistical results?
- A: A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the population parameter with a specified level of confidence (e.g., 95%). It helps assess the precision of an estimate and whether a true effect exists.

Q15: What is the margin of error, and how does it affect the confidence interval?
- A: The margin of error (ME) is the range within which the true population parameter is expected to lie. It is affected by sample size and variability; a smaller ME results in a more precise confidence interval.

Q16: How is Bayes' Theorem used in statistics, and what is its significance?
- A: Bayes' Theorem describes the probability of an event based on prior knowledge of related conditions. It is widely used in statistical inference, decision-making, and machine learning to update beliefs based on new evidence.

Q17: What is the Chi-square distribution, and when is it used?
- A: The Chi-square distribution is used for categorical data analysis. It is commonly applied in tests for goodness-of-fit, independence, and homogeneity to assess how well observed data fit expected distributions.

Q18: What is the Chi-square goodness of fit test, and how is it applied?
- A: The Chi-square goodness-of-fit test evaluates whether an observed frequency distribution matches an expected distribution. It is applied in categorical data analysis to test if data follow a specified distribution.

Q19: What is the F-distribution, and when is it used in hypothesis testing?
- A: The F-distribution is used in statistical tests comparing variances, such as ANOVA and the F-test. It is applied when testing whether multiple population means are equal.

Q20: What is an ANOVA test, and what are its assumptions?
- A: ANOVA (Analysis of Variance) is used to compare means across multiple groups. Its assumptions include:

- The populations are normally distributed.
- Samples are independent.
- Homogeneity of variance (equal variances across groups).

Q21: What are the different types of ANOVA tests?
- A: One-way ANOVA: Compares means across one independent variable with multiple groups.
- Two-way ANOVA: Examines the effect of two independent variables on a dependent variable.
- Repeated measures ANOVA: Used when the same subjects are tested multiple times .

Q22: What is the F-test, and how does it relate to hypothesis testing?
- A: The F-test is a statistical test used to compare variances between groups. It is commonly used in ANOVA to determine if group means are significantly different from each other. The test relies on the F-distribution.

```python
# Q1: Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

# Sample data
sample = np.array([50, 52, 48, 49, 51, 53, 47, 50, 49, 51])
sample_mean = np.mean(sample)
pop_mean = 50  # Known population mean
pop_std = 2  # Population standard deviation
n = len(sample)

# Z-score calculation
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

# P-value calculation (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

```python
# Q4: Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample data
sample = np.array([51, 52, 53, 49, 50, 48, 47, 55, 54, 46])
pop_mean = 50
pop_std = 5
n = len(sample)

# Z-test
sample_mean = np.mean(sample)
z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Visualization
x = np.linspace(-3, 3, 1000)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y, label="Standard Normal Distribution")

# Critical region
z_critical = stats.norm.ppf(0.975)
plt.axvline(z_critical, color="red", linestyle="dashed", label="Critical Value (+1.96)")
plt.axvline(-z_critical, color="red", linestyle="dashed", label="Critical Value (-1.96)")
plt.axvline(z_score, color="blue", linestyle="dashed", label=f"Z-score ({z_score:.2f})")

plt.fill_between(x, y, where=(x > z_critical) | (x < -z_critical), color='red', alpha=0.3)
plt.legend()
plt.title("Two-Tailed Z-Test Decision Region")
plt.show()

print(f"Z-score: {z_score:.3f}, P-value: {p_value:.3f}")
if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Q5: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, beta=0.2):
    x = np.linspace(-3, 3, 1000)
    y_null = norm.pdf(x, 0, 1)
    y_alt = norm.pdf(x, 2, 1)

    plt.plot(x, y_null, label="Null Distribution (H0)")
    plt.plot(x, y_alt, label="Alternative Distribution (H1)", linestyle="dashed")

    z_critical = norm.ppf(1 - alpha)
    plt.axvline(z_critical, color="red", linestyle="dashed", label="Critical Value")

    plt.fill_between(x, y_null, where=(x > z_critical), color="red", alpha=0.3, label="Type I Error (Œ±)")
    plt.fill_between(x, y_alt, where=(x < z_critical), color="blue", alpha=0.3, label="Type II Error (Œ≤)")

    plt.legend()
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.show()

visualize_errors()
```

```python
# Q6: Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy import stats

# Simulated data
group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 5, 30)

# Independent T-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T-score: {t_stat:.3f}, P-value: {p_value:.3f}")

if p_value < 0.05:
    print("Reject the null hypothesis: Significant difference between the groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
# Q9: Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    margin = t.ppf((1 + confidence) / 2, df=n-1) * std_err
    return mean - margin, mean + margin

sample = np.random.normal(50, 5, 30)
ci = confidence_interval(sample)

print(f"95% Confidence Interval: {ci}")
```

```python
# Q10: Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    moe = t.ppf((1 + confidence) / 2, df=n-1) * std_err
    return moe

sample = np.random.normal(50, 5, 30)
moe = margin_of_error(sample)

print(f"Margin of Error: {moe:.3f}")
```

```python
# Q11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, likelihood, marginal):
    return (likelihood * prior) / marginal

# Example: A medical test for a disease
P_disease = 0.01  # Prior: Probability of having the disease
P_positive_given_disease = 0.95  # Likelihood: Test correctly identifies disease
P_positive_given_no_disease = 0.05  # False positive rate
P_positive = (P_positive_given_disease * P_disease) + (P_positive_given_no_disease * (1 - P_disease))

posterior = bayes_theorem(P_disease, P_positive_given_disease, P_positive)
print(f"Posterior Probability: {posterior:.3f}")
```

```python
# Q16: Implement an F-test using Python to compare the variances of two random samples.

group1 = np.random.normal(50, 5, 30)
group2 = np.random.normal(52, 10, 30)

f_stat = np.var(group1, ddof=1) / np.var(group2, ddof=1)
p_value = stats.f.cdf(f_stat, len(group1)-1, len(group2)-1)

print(f"F-statistic: {f_stat:.3f}, P-value: {p_value:.3f}")
```

```python
# Q19: Write a Python function to check assumptions (normality, independence, equal variance) for ANOVA.

def check_anova_assumptions(data_groups):
    for i, data in enumerate(data_groups):
        print(f"Group {i+1} Normality P-value:", stats.shapiro(data)[1])

    f_stat, p_value = stats.levene(*data_groups)
    print(f"Equal Variance P-value: {p_value:.3f}")

check_anova_assumptions([group1, group2, group3])
```

```python
# Q26: Perform a Chi-square test for goodness of fit with simulated data.

import numpy as np
import scipy.stats as stats

# Observed frequencies from survey
observed = np.array([40, 35, 25])

# Expected frequencies assuming equal preference
expected = np.array([100 / 3, 100 / 3, 100 / 3])  # Each category gets ~33.3 customers

# Perform the Chi-square test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Display results
print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation
alpha = 0.05  # Significance level

if p_value < alpha:
    print("Reject the null hypothesis: The customer preferences are NOT equally distributed.")
else:
    print("Fail to reject the null hypothesis: No significant difference in customer preferences.")
```
